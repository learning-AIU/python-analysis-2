### Последовательности

Последовательности — это один из типов данных, поддерживающих оператора проверки на вхождение (```in```), функцию определения размера (```lеn()```), оператора извлечения срезов (```[]```) и возможность выполнения итераций. 

#### Кортежи

Кортеж — это неизменяемая упорядоченная последовательность из нуля и более ссылок на объекты. Его можно создать двумя способами: с помощью функции `tuple()` или заключив элементы в круглые скобки.

````
# Создание кортежей
tp1 = tuple()
tp2 = (1,2,3)

# Количество элементов в кортеже
print(len(tp1))
print(len(tp2))

print(type(tp1))
print(type(tp2))
````

Чтобы обратиться к элементам, можно использовать индексы. Не забываем, что индексация начинается с нуля. Отрицательная индексация означает, что мы ведем отсчет с конца.

````
# Позиции элементов в кортеже
print(tp2[0], tp2[1], tp2[2])
print(tp2[-3], tp2[-2], tp2[-1])
````

Также в кортежах можно использовать срезы. Они используются так же, как в строках. Срезы указываем в квадратных скобках, где прописываем 3 числа, разделенные двоеточиями:

- первое — начальный индекс;
- второе — конечный индекс + 1 (он не включается в срез);
- третье — шаг, с которым мы отбираем элементы.

Эти параметры могут быть как положительными, так и отрицательными (отрицательный индекс — это номер элемента с конца) и использоваться по отдельности. Однако все равно нужно использовать двоеточие, чтобы Python понимал, какой именно параметр мы хотим задать.

````
# Операция среза в кортеже
print(tp2[:])
print(tp2[-3:])
print(tp2[0:-1])
print(tp2[0:2])
print(tp2[::2])
print(tp2[::-1])
````

Кортежи — это неизменяемый тип данных, поэтому мы не можем изменить их элементы.

````
# Проверка неизменяемости
tp2[0]=1
````

Если функция возвращает несколько значений через занятую, эти элементы возвращаются в виде кортежа:

````
#  Работа с функциями
def f_sum_prod(a, b):
  return a + b, a * b

f_sum_prod(2, 5)
````

Еще мы можем поэлементно сопоставить элементы из двух кортежей с помощью `zip`:

````
# Использование функции zip
a = (1, 2, 3, 4)
b = (5, 6, 7, 8)
for i in zip(a, b):
  print(i)
````

#### Именованные кортежи

В `collections` есть также именованные кортежи, которые необходимы, когда мы хотим понимать, какую информацию несет в себе каждый элемент кортежа:

````
import collections
dir(collections)
````

Для начала создаем прототип именованного кортежа, где задаем имена каждого элемента:

````
# Создание прототипа именованного кортежа
User = collections.namedtuple('User','name email', 'reg_date')
````

А затем с помощью него создаем кортеж:

````
# Кортеж из пользователей
users_list = (User('Boris','b@gmail.com','01.02.00'),
              User('Anton','a@gmail.com','11.02.01'),
              User('Don','d@gmail.com','01.04.02'))
````

````
users_list
````

Так мы можем интуитивно понятнее и информативнее обратиться к элементам по имени полей:

````
# Обращение к записям
print(users_list[0].name)
print(users_list[0].reg_date)
````

Но возможность обратиться по индексам все еще остается:

````
# Пример удобства в чтении кода
Iam = User('Alex','al@gmail.com','01.02.02')
print(Iam.name, Iam[0])
print(Iam.email, Iam[1])
print(Iam.reg_date, Iam[2])
````

#### Списки

Список — это изменяемая упорядоченная последовательность из нуля и более ссылок на объекты. В них мы можем изменять элементы, добавлять и удалять их.

Методы списков приведены в таблице:
![alt text](https://drive.google.com/uc?id=1_HbvqT7omF7rxmtZ3PHLzeoeqxxccVM2)

Рассмотрим, как применять некоторые из них:

````
# Создание списков
list1 = list()
list2 = []
list3 = [1,2,3,4]
````

````
# Добавление элементов в списки
list1.append(1)
print(list1)
````

````
# Добавление списка в качестве элемента списка
list1.append(list3)
print(list1)
````

````
# Конкатенирование двух списков
list1 += list3
print(list1)
````

К элементам списка мы можем также обращаться по индексам и брать их срезы. Принцип работы такой же, как в строках и кортежах:

````
# Индексация списка
print(list1[0])
print(list1[1])
print(list1[5])
print(list1[1][2])
````

````
# Операция среза
print(list1[:])
print(list1[1:])
print(list1[1][2:])
print(list1[::-1])
````

Удалять элементы можно двумя способами: используя `remove` или `del`. В первом случае мы передаем значение, которое хотим удалить, и он удаляет первый элемент, который встречается с таким значением:

````
# Удаление элементов списка
list1.remove(2)
print(list1)
````

`del` удаляет сам объект, поэтому мы должны передать именно ссылку на объект:

````
del list1[0]
print(list1)
````

````
del list1[0][3]
print(list1)
````

Мы можем вставить в определенное место списка любой элемент, даже другой список, но это довольно опасно. Списки хранят ссылки на объекты. Если добавить в него ссылку на другой список, то при изменении одного из них изменится и второй.

````
# Вставка элемента
list1.insert(0, list3)
print(list1)
````

````
# В list3 косвенно был удален элемент, поскольку list1[0][3] — это ссылка на элемент в list3
list3
````

````
# а при изменении list3 изменяется и элемент list1
list3[0] = 999
print(list1)
````

Пока это может показаться сложным и запутанным, но позже вы к этому привыкните.

#### Множества

Множества — это коллекции, которые содержат в себе только уникальные элементы.

Рассмотрим только тип ```set``` — неупорядоченную коллекцию из нуля и более ссылок различных объектов. Поддерживает операции на множествах (в математическом смысле): объединение, пересечение, разность и т. д.

Множество можно создать двумя способами: с помощью функции `set` или фигурных скобок. **Важно: пустые скобки создают не множество, а словарь.**

````
# Создание множества
st1 = set()
st2 = {}
st3 = {1,2,3,3,4,2}
st4 = {1,4,6,8}
````

````
# Пустые {} создают не множество set!
print(type(st1))
print(type(st2))
````

Можно заметить, что в множестве `set3` при создании было указано две 2, но хранит он только одну.

````
# Одно из возможных использований — получение множества уникальных значений
print(st3)
````

Мы можем добавить элемент в множество. Если он там уже есть, то множество просто не изменится.

````
# Добавление элементов в множество
st3.add(0)
print(st3)
````

У множеств есть операции, которые свойственны только им:

Но так как элементы множества не упорядочены, мы не можем обратиться к ним по индексу.

````
# Отсутствие упорядоченности
st3[0]
````

### Отображения

#### Словари

Тип ```diсt``` — это неупорядоченная коллекция из нуля или более пар
«ключ-значение», в которых в качестве ключей могут использоваться
ссылки на неизменяемые (хешируемые) объекты, а в качестве значений —
ссылки на объекты любого типа.

![alt text](https://drive.google.com/uc?id=1Wy8CdeDiKaPu1wKLDT7rYdOGv24lteVi)

Создать словарь можно двумя способами: с помощью функции `dict` или фигурных скобок:

````
# Создание словаря
dict1 = {}
dict2 = dict()
dict3 = {'a':1, 'b':2, 4:[1,2], (1,'Ted'):[(1,2),1,'abc']}
dict4 = dict.fromkeys(['a','b'],0)
dict5 = dict(zip(['a','b'],[0,1]))
````

````
print(dict3)
print(dict4)
print(dict5)
````

Мы можем расширить один словарь элементами другого с помощью `update`:

````
# Добавление значений
dict5.update({'c':3})
print(dict5)
````

В словарях нет индексации. Вместо этого мы можем получить значение словаря, обратившись по его ключу. В качестве ключа может служить любой неизменяемый тип данных (число, строка, кортеж и т. д.). Ключи в словаре должны быть уникальными. Если мы попытаемся записать значение в существующий ключ, мы потеряем значение, которое хранилось в нем до этого.

````
# Индексация
print(dict3['a'])
print(dict3['b'])
print(dict3[(1,'Ted')])
print(dict3[4])
````

````
# Проверка упорядоченности
print(dict3[0])
````

Мы можем получить список всех значений, ключей или пар «ключ-значение» с помощью соответствующих методов: `values`, `keys`, `items`:

````
# Просмотр всех значений, ключей, пар «ключ-значение»
print(dict3.values())
print(dict3.keys())
print(dict3.items())
````
